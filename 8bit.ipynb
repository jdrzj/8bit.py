```python
import numpy as np
import matplotlib.pyplot as plt
# source: https://www.pexels.com/photo/husky-and-akita-dogs-sitting-together-outdoors-7564097/
im = np.array(Image.open('doggos.jpg'))
im_avg = im
im_median = im
im_mean = im
plt.imshow(im)
```

```python
pix_size = 50
(h, w, _) = im_avg.shape
parts_x = int(h/pix_size)+1
parts_y = int(w/pix_size)+1
[parts_x, parts_y]

for x in range(0, parts_x):
    for y in range(0, parts_y):
        box = im[(x*pix_size):((x+1)*pix_size), (y*pix_size):((y+1)*pix_size)]
        avg = np.average(box, axis=(0,1))
        mean = np.mean(box, axis=(0,1))
        med = np.median(box, axis=(0,1))
        im_avg[(x*pix_size):((x+1)*pix_size), (y*pix_size):((y+1)*pix_size)] = avg
        im_mean[(x*pix_size):((x+1)*pix_size), (y*pix_size):((y+1)*pix_size)] = mean
        im_median[(x*pix_size):((x+1)*pix_size), (y*pix_size):((y+1)*pix_size)] = med

fig = plt.figure(figsize=(30,30))
    
fig.add_subplot(1, 3, 1)
plt.imshow(im_avg)
fig.add_subplot(1, 3, 2)
plt.imshow(im_mean)
fig.add_subplot(1, 3, 3)
plt.imshow(im_median)
im_avg = Image.fromarray(im_avg)
im_avg.save("doggos_8bit.jpg")
plt.show()
```
